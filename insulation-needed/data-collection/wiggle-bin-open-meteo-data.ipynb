```python
import requests
import pandas as pd
```

```python
# --- USER SETTINGS ---
lat = 51.8425   # Nijmegen approx.
lon = 5.8528
start = "2021-01-01"
end = "2021-12-31"
output_csv = "weather_2021.csv"
```

```python
# --- 1. Fetch weather data from Open-Meteo ---
url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={lat}&longitude={lon}"
    f"&start_date={start}&end_date={end}"
    f"&hourly=temperature_2m&timezone=Europe/Amsterdam"
)
resp = requests.get(url)
resp.raise_for_status()
data = resp.json()
times = pd.to_datetime(data["hourly"]["time"])  # parse as datetime
temps = [float(t) for t in data["hourly"]["temperature_2m"]]

weather_df = pd.DataFrame({
    "timestamp": times,
    "T_out": temps
})

# --- 3. Save to CSV ---
weather_df.to_csv(output_csv, index=False)
print(f"Weather data saved to {output_csv}")
```
